```python
Ha = Alex.flux_halpha
Hb = Alex.flux_hbeta
OIII = Alex.flux_oiii
OII = Alex.flux_oii
```

```python
Ha_num =[]
for i in range(len(Ha)):
    try:
        Ha_num.append(float(Ha[i]))
    except:
        Ha_num.append(float(-99))
```

```python
Hb_num =[]
for i in range(len(Hb)):
    try:
        Hb_num.append(float(Hb[i]))
    except:
        Hb_num.append(float(-99))
```

```python
OIII_num =[]
for i in range(len(OIII)):
    try:
        OIII_num.append(float(OIII[i]))
    except:
        OIII_num.append(float(-99))
```

```python
OII_num =[]
for i in range(len(OII)):
    try:
        OII_num.append(float(OII[i]))
    except:
        OII_num.append(float(-99))
```

```python
Ha_num = np.array(Ha_num)
```

```python
type(Alex.flux_hbeta[0])
```

```python
plt.hist(np.log10(Ha_num[Ha_num>0.]),bins=30,histtype='step',lw=3)
ha_med = np.median(np.log10(Ha_num[Ha_num>0.]))
plt.axvline(ha_med)
plt.title(ha_med)
plt.xlabel(r'$H_{\alpha}$')
plt.show()
```
