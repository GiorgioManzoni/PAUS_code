```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
caller = pd.read_table('../data/CFHTLens_lotsofcolumns_stars.tsv')
```

```python
other = pd.read_table('../data/CFHTLens_MASK_only.tsv')
```

```python
LENS_with_MASK = caller.set_index('id').join(other.set_index('id'))
```

```python
LENS_with_MASK.to_csv('../data/LENS_all_MASK.csv')
```

```python
plt.figure(figsize=(12,10))
Mmask = np.array(M)[np.where(M['MASK']==1)]
plt.plot(M['ALPHA_J2000'],M['DELTA_J2000'],'.',markersize=6,label='ALL')
plt.plot(Mmask['ALPHA_J2000'],Mmask['DELTA_J2000'],'.',markersize=6,label='MASK')
plt.legend(markerscale=4.,loc=1,framealpha=1)
plt.xlim(214.5,215)
plt.ylim(54.5,55)
plt.xlabel('RA J2000')
plt.ylabel('DEC J2000')
plt.title('W3 field + Zoom in')
#plt.savefig('../plots/14/w3w3_lens_ls_zoom_plus.png')
plt.show()
```
